## User Overlap Investigation

Before selecting a final dataset, I want to determine whether the two Fitbit exports contain the same participants. If user IDs overlap and the date ranges are consecutive, the datasets may be combined to increase the analysis period and sample size.

```python
df1 = pd.read_csv(export1 / "dailyActivity_merged.csv")
df2 = pd.read_csv(export2 / "dailyActivity_merged.csv")

users1 = set(df1["Id"].unique())
users2 = set(df2["Id"].unique())

print("Users in Export 1:", len(users1))
print("Users in Export 2:", len(users2))

overlap = users1.intersection(users2)

print("Shared Users:", len(overlap))
```

```python
print("\nExport 1 IDs:")
print(sorted(users1))

print("\nExport 2 IDs:")
print(sorted(users2))
```

## Preliminary Findings: User Overlap

The initial comparison of participant IDs revealed substantial overlap between the two Fitbit exports.

### Key Findings

- Export 1 contains data from 35 unique users.
- Export 2 contains data from 33 unique users.
- All 33 users present in Export 2 also appear in Export 1.
- Two users (2891001357 and 6391747486) appear only in Export 1 and do not continue into Export 2.

### Initial Interpretation

The results suggest that the two exports likely represent consecutive collection periods from the same Fitbit study rather than two independent datasets.

Because nearly all participants appear in both exports, combining the datasets may provide a longer observation period and a more complete view of user behavior. However, before making a final dataset selection decision, the collection dates must be examined to determine whether the exports are consecutive and whether any overlapping records exist.

```python
print("Export 1 Date Range")
print("Start:", df1["ActivityDate"].min())
print("End:", df1["ActivityDate"].max())

print("\nExport 2 Date Range")
print("Start:", df2["ActivityDate"].min())
print("End:", df2["ActivityDate"].max())
```

## Preliminary Findings: Date Range Comparison

The date ranges of the two Fitbit exports were compared to determine whether the datasets could be combined for analysis.

### Key Findings

- Export 1 contains records from March 12, 2016 through April 9, 2016.
- Export 2 contains records from April 12, 2016 through May 9, 2016.
- The exports do not overlap.
- A two-day gap exists between the datasets (April 10-11, 2016).

### Interpretation

The results indicate that the two exports represent consecutive phases of the same Fitbit study. Although a small two-day gap exists between the collection periods, the participant overlap and continuity of dates suggest that combining the datasets will provide a more complete view of user behavior than analyzing either export independently.

### Preliminary Decision

The two exports will be merged during the data preparation phase. The two-day gap will be documented as a limitation but is not expected to materially impact the analysis.

## Dataset Selection Decision

The two Fitbit exports were evaluated to determine whether they should be analyzed independently or merged into a single dataset.

### Evidence Supporting a Merge

- Export 1 contains 457 records across 35 users.
- Export 2 contains 940 records across 33 users.
- All users present in Export 2 are also present in Export 1.
- Only two users appear exclusively in Export 1.
- The datasets cover consecutive collection periods with only a two-day gap.
- Both datasets share the same schema, column structure, and data types.

### Decision

The two exports will be merged into a single master dataset for analysis.

### Rationale

Combining the datasets increases the observation period from approximately one month to nearly two months while maintaining participant continuity. This provides a more complete representation of user behavior and strengthens the reliability of subsequent analyses.

### Limitation

A two-day gap exists between April 9, 2016 and April 12, 2016. This gap will be documented but is not expected to materially affect the analysis.
